```python
from deep_translator import GoogleTranslator
from gtts import gTTS
from IPython.display import display, Audio
import ipywidgets as widgets

# Language Dictionary
languages = {
    "English": "en",
    "Hindi": "hi",
    "Bengali": "bn",
    "Spanish": "es",
    "French": "fr",
    "German": "de",
    "Chinese": "zh-CN",
    "Japanese": "ja",
    "Arabic": "ar"
}

# UI Elements
text_input = widgets.Textarea(
    placeholder='Enter text here...',
    description='Input:',
    layout=widgets.Layout(width='100%', height='100px')
)

source_dropdown = widgets.Dropdown(
    options=languages.keys(),
    value="English",
    description='Source:'
)

target_dropdown = widgets.Dropdown(
    options=languages.keys(),
    value="Hindi",
    description='Target:'
)

translate_button = widgets.Button(
    description="Translate",
    button_style='success'
)

output_area = widgets.Output()

# Translation Function
def translate_text(b):
    output_area.clear_output()

    text = text_input.value
    src_lang = languages[source_dropdown.value]
    dest_lang = languages[target_dropdown.value]

    if text.strip() == "":
        with output_area:
            print("âš  Please enter some text.")
        return

    try:
        translated = GoogleTranslator(source=src_lang, target=dest_lang).translate(text)

        with output_area:
            print("âœ… Translated Text:\n")
            print(translated)

            # Text-to-Speech
            tts = gTTS(translated, lang=dest_lang)
            tts.save("translated_audio.mp3")

            print("\nðŸ”Š Audio:")
            display(Audio("translated_audio.mp3"))

    except Exception as e:
        with output_area:
            print("âŒ Error:", e)

translate_button.on_click(translate_text)

# Display App
display(text_input)
display(source_dropdown)
display(target_dropdown)
display(translate_button)
display(output_area)
```
